```python
import pandas as pd
import numpy as np
import seaborn as sns # 데이터 시각화를 위한 seaborn 라이브러리 임포트
import matplotlib.pyplot as plt # 데이터 시각화를 위한 matplotlib의 pyplot 모듈 임포트
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression  # 선형 회귀 모델을 위한 LinearRegression 임포트
from sklearn.preprocessing import StandardScaler  # 데이터 정규화를 위한 StandardScaler 임포트
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  # 모델 평가를 위한 metrics 임포트



# 데이터셋 불러오기 (파일 경로 지정)
file_path = r"C:\Users\gram15\Desktop\python_basic_10\python_basic_1\machine.data_update.csv" 
columns = ["vendor_name", "model_name", "MYCT", "MMIN", "MMAX", "CACH", "CHMIN", "CHMAX", "PRP", "ERP"] # 데이터 파일 열 이름 
data = pd.read_csv(file_path, names=columns) #데이터셋을 pandas 데이터프레임으로 읽어오기

# 숫자형 데이터만 선택하고, 데이터 프레임 생성
numeric_data = data[["MYCT", "MMIN", "MMAX", "CACH", "CHMIN", "CHMAX", "PRP"]].apply(pd.to_numeric, errors='coerce')

# 결측값 확인, 결측값 제거
print(numeric_data.isnull().sum()) # 결측값 개수 출력
numeric_data = numeric_data.dropna() # 결측값이 존재하는 행 제거

# 데이터 탐색
print(numeric_data.head()) # 데이터프레임의 첫 5행 출력
print(numeric_data.describe()) # 데이터프레임의 통계 요약 출력
print(numeric_data.info()) # 데이터프레임의 정보 출력

# 타겟 변수 PRP 히스토그램
plt.figure(figsize=(10, 6))
sns.histplot(numeric_data['PRP'], kde=True)  #PRP 분포에 대한 히스토그램 및 KDE 그래프
plt.title('Distribution of real values') # 그래프 제목 설정
plt.show()

# 특징 변수 간의 상관 관계
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm') # 상관행렬 표시
plt.title('correlation matrix')
plt.show()

# 특징과 타겟 변수 정의
features = numeric_data[["MYCT", "MMIN", "MMAX", "CACH", "CHMIN", "CHMAX"]] # 특징 변수
target = numeric_data["PRP"] # 타겟 변수

# 데이터 정규화
scaler = StandardScaler() # StandardScaler 객체 생성
features_scaled = scaler.fit_transform(features) # 특징 변수 정규화

# 훈련 및 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42) # 데이터 분할

# 선형 회귀 모델 훈련
model = LinearRegression()  # LinearRegression 객체 생성
model.fit(X_train, y_train)  # 모델 훈련

# 교차 검증
cross_val_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2') # 교차 검증 수행
print(f"Cross-validation R^2 scores: {cross_val_scores}") # 교차 검증 R^2 점수 출력
print(f"Mean cross-validation R^2 score: {np.mean(cross_val_scores)}")  # 평균 교차 검증 R^2 점수 출력

# 훈련 데이터 성능 평가
y_train_pred = model.predict(X_train)  # 훈련 데이터에 대한 예측값 생성
train_mse = mean_squared_error(y_train, y_train_pred) # 훈련 데이터 MSE 계산
train_mae = mean_absolute_error(y_train, y_train_pred) # 훈련 데이터 MAE 계산
train_r2 = r2_score(y_train, y_train_pred) # 훈련 데이터 R^2 점수 계산

print(f"Training Data - Mean Squared Error (MSE): {train_mse}")  # 훈련 데이터 MSE 출력
print(f"Training Data - Mean Absolute Error (MAE): {train_mae}") # 훈련 데이터 MAE 출력
print(f"Training Data - R^2 Score: {train_r2}") # 훈련 데이터 MAE 출력

# 테스트 데이터 성능 평가
y_test_pred = model.predict(X_test) # 테스트 데이터에 대한 예측값 생성
test_mse = mean_squared_error(y_test, y_test_pred)  # 테스트 데이터 MSE 계산
test_mae = mean_absolute_error(y_test, y_test_pred) # 테스트 데이터 MSE 계산
test_r2 = r2_score(y_test, y_test_pred) # 테스트 데이터 R^2 점수 계산

print(f"Test Data - Mean Squared Error (MSE): {test_mse}") # 테스트 데이터 MSE 출력
print(f"Test Data - Mean Absolute Error (MAE): {test_mae}") # 테스트 데이터 MAE 출력
print(f"Test Data - R^2 Score: {test_r2}") # 테스트 데이터 R^2 점수 출력

#  훈련 데이터 실제 값과 예측 값 시각화 
plt.figure(figsize=(10, 6)) # 실제 값 vs 예측 값 산점도 (훈련 데이터)
plt.scatter(y_train, y_train_pred, edgecolors=(0, 0, 0)) # 실제 값 vs 예측 값 산점도
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], 'r--', lw=2)
plt.xlabel("actual value")  # x축 레이블 설정
plt.ylabel("predicted value")  # y축 레이블 설정
plt.title("Training data: actual vs predicted values")  # 그래프 제목 설정
plt.show()  # 그래프 출력

# 테스트 데이터 실제 값과 예측 값 시각화 
plt.figure(figsize=(10, 6)) # 실제 값 vs 예측 값 산점도 (테스트 데이터)
plt.scatter(y_test, y_test_pred, edgecolors=(0, 0, 0))
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--', lw=2)
plt.xlabel("actual value") # x축 레이블 설정
plt.ylabel("predicted value ")  # y축 레이블 설정
plt.title("Test data: actual vs predicted values") # 그래프 제목 설정
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score # 데이터 분할 및 교차 검증 sklearn 모듈 임포트
from sklearn.preprocessing import StandardScaler # 데이터 정규화를 위한 StanddardScler 임포트
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor # 앙상블 모델 임포트
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score # 성능 평가 임포트

# 데이터셋 불러오기
file_path = r"C:\Users\gram15\Desktop\python_basic_10\python_basic_1\machine.data_update.csv" # 경로 설정
columns = ["vendor_name", "model_name", "MYCT", "MMIN", "MMAX", "CACH", "CHMIN", "CHMAX", "PRP", "ERP"]
data = pd.read_csv(file_path, names=columns)

# 숫자형 데이터만 선택하여 새로운 데이터프레임 생성
numeric_data = data[["MYCT", "MMIN", "MMAX", "CACH", "CHMIN", "CHMAX", "PRP"]].apply(pd.to_numeric, errors='coerce')

# 결측값 확인 및 제거
numeric_data = numeric_data.dropna() # 결측값이 있는 행 제거 

# 특징과 타겟 변수 정의
features = numeric_data[["MYCT", "MMIN", "MMAX", "CACH", "CHMIN", "CHMAX"]] # 특징 변수 선택
target = numeric_data["PRP"] # 타겟 변수 선택
# 데이터 정규화
scaler = StandardScaler() # StandardScaler 객체 생성
features_scaled = scaler.fit_transform(features) # 특징 데이터 정규화

# 훈련 및 테스트 데이터 분할하기
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)

# 랜덤 포레스트 모델 훈련
rf_model = RandomForestRegressor(n_estimators=100, random_state=42) # 랜덤 포레스트 모델 생성
rf_model.fit(X_train, y_train) # 모델 훈련

# Gradient Boosting 모델 훈련
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42) #  Gradient Boosting 모델 생성
gb_model.fit(X_train, y_train) # 모델 훈련

# 교차 검증
rf_cross_val_scores = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='r2') # 랜덤 포레스트 모델의 교차 검증 R^2 점수 계산
gb_cross_val_scores = cross_val_score(gb_model, X_train, y_train, cv=5, scoring='r2') # Gradient Boosting 모델의 교차 검증 R^2 점수 계산

print(f"Random Forest - Cross-validation R^2 scores: {rf_cross_val_scores}") # 랜덤 포레스트 모델의 교차 검증 R^2 점수 출력
print(f"Random Forest - Mean cross-validation R^2 score: {np.mean(rf_cross_val_scores)}") # 랜덤 포레스트 모델의 평균 교차 검증 R^2 점수 출력
print(f"Gradient Boosting - Cross-validation R^2 scores: {gb_cross_val_scores}") # Gradient Boosting 모델의 교차 검증 R^2 점수 출력
print(f"Gradient Boosting - Mean cross-validation R^2 score: {np.mean(gb_cross_val_scores)}") # Gradient Boosting 모델의 평균 교차 검증 R^2 점수 출력
# 랜덤 포레스트 모델 성능 평가 (훈련 데이터)
y_train_pred_rf = rf_model.predict(X_train) # 훈련 데이터 예측값 생성
train_mse_rf = mean_squared_error(y_train, y_train_pred_rf) # MSE 계산
train_mae_rf = mean_absolute_error(y_train, y_train_pred_rf) # MAE 계산
train_r2_rf = r2_score(y_train, y_train_pred_rf) # R^2 점수 계산

print(f"Random Forest - Training Data - Mean Squared Error (MSE): {train_mse_rf}") # 테스트 데이터에 대한 MSE 계산
print(f"Random Forest - Training Data - Mean Absolute Error (MAE): {train_mae_rf}") # 테스트 데이터에 대한 MAE 계산
print(f"Random Forest - Training Data - R^2 Score: {train_r2_rf}") # 테스트 데이터에 대한 R^2 점수 계산


# 랜덤 포레스트 모델 성능 평가 (테스트 데이터)
y_test_pred_rf = rf_model.predict(X_test) # 테스트 데이터 예측값 생성
test_mse_rf = mean_squared_error(y_test, y_test_pred_rf) # MSE 계산
test_mae_rf = mean_absolute_error(y_test, y_test_pred_rf) # MAE 계산
test_r2_rf = r2_score(y_test, y_test_pred_rf) # R^2 점수 계산

print(f"Random Forest - Test Data - Mean Squared Error (MSE): {test_mse_rf}")  # MSE 출력
print(f"Random Forest - Test Data - Mean Absolute Error (MAE): {test_mae_rf}") # MAE 출력
print(f"Random Forest - Test Data - R^2 Score: {test_r2_rf}") # R^2 점수 출력

# Gradient Boosting 모델 성능 평가 (훈련 데이터)
y_train_pred_gb = gb_model.predict(X_train)
train_mse_gb = mean_squared_error(y_train, y_train_pred_gb)
train_mae_gb = mean_absolute_error(y_train, y_train_pred_gb)
train_r2_gb = r2_score(y_train, y_train_pred_gb)

print(f"Gradient Boosting - Training Data - Mean Squared Error (MSE): {train_mse_gb}")
print(f"Gradient Boosting - Training Data - Mean Absolute Error (MAE): {train_mae_gb}")
print(f"Gradient Boosting - Training Data - R^2 Score: {train_r2_gb}")

# Gradient Boosting 모델 성능 평가 (테스트 데이터)
y_test_pred_gb = gb_model.predict(X_test)
test_mse_gb = mean_squared_error(y_test, y_test_pred_gb)
test_mae_gb = mean_absolute_error(y_test, y_test_pred_gb)
test_r2_gb = r2_score(y_test, y_test_pred_gb)

print(f"Gradient Boosting - Test Data - Mean Squared Error (MSE): {test_mse_gb}")
print(f"Gradient Boosting - Test Data - Mean Absolute Error (MAE): {test_mae_gb}")
print(f"Gradient Boosting - Test Data - R^2 Score: {test_r2_gb}")

# 실제 값과 예측 값 시각화 (랜덤 포레스트, 훈련 데이터)
plt.figure(figsize=(10, 6)) # 그래프 크키 설정
plt.scatter(y_train, y_train_pred_rf, edgecolors=(0, 0, 0)) # 실제 값과 예측 값 산점
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], 'r--', lw=2) # 실제 값과 예측 값의 1:1 
plt.xlabel("actual values") # x축라벨
plt.ylabel("predicted values") # y축 라벨
plt.title("Random Forest - Training Data: actual vs predicted values")
plt.show() # 출력하기

# 실제 값과 예측 값 시각화하기 (랜덤 포레스트, 테스트 데이터)
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_test_pred_rf, edgecolors=(0, 0, 0))
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--', lw=2)
plt.xlabel("actual values")
plt.ylabel("predicted values")
plt.title("Random Forest - Test Data: actual vs predicted values")
plt.show()

# 실제 값과 예측 값 시각화하기  (Gradient Boosting, 훈련 데이터)
plt.figure(figsize=(10, 6))
plt.scatter(y_train, y_train_pred_gb, edgecolors=(0, 0, 0))
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], 'r--', lw=2)
plt.xlabel("actual values")
plt.ylabel("predicted values")
plt.title("Gradient Boosting - Training Data: actual vs predicted values")
plt.show()

# 실제 값과 예측 값 시각화하기 (Gradient Boosting, 테스트 데이터)
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_test_pred_gb, edgecolors=(0, 0, 0))
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--', lw=2)
plt.xlabel("actual values")
plt.ylabel("predicted values")
plt.title("Gradient Boosting - Test Data: actual vs predicted values ")
plt.show()
```
